## **`Introduction to Functions`**

**A function is a reusable code block that performs a specific task.**

🧠 Think of it like a machine:

- You give it some input

- It does something

- It gives back output


💡 so, Function is kind of like a machine (let’s say a coffee machine) which can take input (e.g., coffee beans, milk, sugar, etc.), then process it (grind the beans, heat the water, mix everything together), and finally give some output (a cup of coffee).

- In Python, functions are defined using the `def` keyword.

- After def, we write the function name, followed by parentheses `()` and a colon `:`.

- The code inside the function must be `indented (spaced)`.

- Running a function is called `function calling`.

- Basic Syntax
def function_name():
    codeblock


````python
#Creating a function that greets with people.

def greet(name = "Zubayer"): # setting default value of parameters is a must.
  print(f"Hello,{name}.How's going?")

greet()
greet("Sadia")
````

````python
#Another example with return value
def greet(name = "Zubayer"):
    print(f"Hello {name}")
    print(f"How are you {name}?")
    return # return function gives access to use the variable outside the function block, but Print doesn't.

greet()
greet("Noman")
````

## Parameters & Arguments

`Parameters` are variables defined inside the function parentheses.
They act as placeholders so the function can work with different values.

`Arguments` are the actual values you pass into the function when calling it.

#### 📌 Why This Matters
Parameters allow your functions to work with **dynamic data**, making them flexible and reusable.

#### Example

```python
def greet_name(name):
    print(f"Hello {name}")
```

Here:
- `name` → Parameter
- It is defined when we create the function.

#### Calling (Running) the Function
```python
greet_name("Sayem")
greet_name("Shovon")
```

Here:
- `Sayem` and `Shovon` → Arguments
- These are the real values passed into the function.


**Default parameter:** A default parameter is a parameter that already has a value assigned in the function definition.

If you don’t provide a value when calling the function, Python uses the default.

```python
# Example
def greet_name(name="Guest"):
    print(f"Hello {name}")

greet_name("Sayem")   # Output: Hello Sayem
greet_name()          # Output: Hello Guest
```

- "Sayem" → overrides the default

- No argument → uses "Guest" as the default

##Positional Arguments  and Keyword Arguments
🏷️ Positional Arguments
In it, you have to remeber the parametes order that you put in a function while defining it. if you do not maitain position when you put arguments you will get wrong results or an error.

🏷️ Keyword Arguments

Explicitly specify parameter names — order does not matter.

These features make functions easier to use and harder to misuse.

````python
#For positional arguments
def calculate_final_price(price, tax_rate):
    total = price + (price * tax_rate)
    print(f"Total: {total}")
    return total

calculate_final_price(0.15, 3000)   # wrong argument order-> 0.15 (tax_rate) should be the 2nd argument
calculate_final_price(3000, 0.15)   # right argument order
````

````python
def send_email(user_email, subject="Welcome!", body="Thanks for joining."):
    print(f"To: {user_email}")
    print(f"Subject: {subject}")
    print(f"Content: {body}")

send_email("mdzubayereashraf@gmail.com") # user_email must be passed - no default value for user_email
````

````python
 #For keyword arguments
def calculate_final_price(price, tax_rate):
    total = price + (price * tax_rate)
    print(f"Total: {total}")
    return total
# postion doesn't matter  -> keyword arguments
calculate_final_price(price=3000, tax_rate=0.15)
calculate_final_price(tax_rate=0.15, price=3000)
````

## Variable Scope

`Scope` means where a variable can be accessed in your code.


1. Local Scope: A variable created inside a function
```python
def greet():
    message = "Hello"
    print(message)

greet()    ##  works
print(message) ## Error
```
- variable `message` works inside the function
- It does NOT exist outside
- Local variables live only inside the function


2. Global Scope: A variable created outside all functions.

```python
name = "Shovon"

def show():
    print(f"Hello {name}")

show()   ## works
print(name)  ## also works
```

- Variable `name` is accessible both inside and outside of the function
- Global variables can be read inside functions.

#### 🏠 Think of Scope Like Rooms in a House

- **Local Scope**: Variables inside a function (private room)
- **Global Scope**: Variables outside functions (shared space)

Understanding scope prevents **unexpected bugs**.

````python
def greet():
    user = "Sayem"   # local scope
    print(f"Hello, welcome back {user}")
greet()
print(user) # here variable `user` cannot be accessed outside of the function
````

````python
user_name = "Jahid"  # global scope

def greet_v2():
    print(f"Hello, welcome back {user_name}")

greet_v2()
print(user_name) # here variable `user_name` can be accessed outside of the function
````
