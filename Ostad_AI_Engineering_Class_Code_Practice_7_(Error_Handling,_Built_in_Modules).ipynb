## Error Handling with `Try & Except`
Errors happen. Good software handles them gracefully instead of crashing.

🎯 Why Use Try-Except?
- Protect your program from invalid user input
- Improve user experience
- Essential for production-ready code
- Protect your program from crashing

```python
# What happens when users put invalid input?
def division(a,b):
  if b == 0:
    print("Division is not possible.")
  else :
      return a / b
division(30,5)
division(0,5)
division(5,0)
division("a",5) # if user put wrong input accidentally ,program crashes.
```

```python
#We can handle this kinna problem using try and except commands
def safe_division(a, b):
   try:
     if b == 0:
       return"Division is not possible."
     result = a / b
     return result

   except Exception as e: # this command shows the errors without mentioning error type.
       return e

print(safe_division(30,5))

print(safe_division(0,5))

print(safe_division(5,0))

print(safe_division(5,"v"))

```

## `custom user friendly messages for each error`

```python
def safe_division(a,b):
  try:
    x = a / b
    return x
  except TypeError:
    print("Error: Please provide numeric values only.")
  except ZeroDivisionError:
    print("Error: You cannot divide by zero!")

print(safe_division(10, 2))
print(safe_division(10, 0))
print(safe_division(10, "a"))
```

## `A single custom error message for all kinna errors`

```python
def safe_division(a,b):
  try:
    x = a / b
    return  x
  except Exception: # Exception is used to show a common customize massage for every single error.
    print("Sorry something went wrong .Please try again, buddy.")

print(safe_division(10, 2))
print(safe_division(10, 0))
print(safe_division(10, "a"))
```
