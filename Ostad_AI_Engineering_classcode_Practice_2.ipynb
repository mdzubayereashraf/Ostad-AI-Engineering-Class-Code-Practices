**String Indexing**

````python
#Positive Indexing...Varibalename[Index number]
text1 = "Zubayer"
print(text1[0]) # Z
print(text1[1]) # u
print(text1[2]) # b
print(text1[3]) # a
print(text1[4]) # y
print(text1[5]) # e
````

````python
#Negative indexing ....Varibalename[Index number]
text2 = "Zubayer"
print(text2[-1]) # r
print(text2[-2]) # y
print(text2[-3]) # b
print(text2[-4]) # a
print(text2[-5]) # u
print(text2[-6]) # Z
````

**String Slicing**

````python
# variable name[start:stop:sep]
text = "Zubayer"
print(text[1:4])
print(text[:4])
print(text[2:])
print(text[::2])# only used seperator
print(text[::-1])# only used seperator
````

## **Conditional Statements**

Python conditional statements are used to execute specific blocks of code based on whether a condition is true or false, allowing programs to make decisions and control their flow.

The main keywords are:

`if` → check a condition

`elif` → check another condition if the previous `if`/`elif` was False

`else` → run code when none of the conditions are True

**N.B.** A condition → is either `True` or `False`





**Basic IF Statement**


````python
age = int(input("Enter your age: "))
if age >= 18 :
  print("you are an adult")
````

**If - else Statement**

````python
age = int(input("Enter your age: "))

if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
````

**if-elif-else statement**

````python
age = int(input("Enter your age: "))

if age < 13:
    print("You are a child.")
elif age < 20:
    print("You are a teenager.")
else:
    print("You are an adult.")
````

**Multiple elif Statement**

````python
age = int(input("Enter your age: "))

if age < 13:
    print("You are a child.")
elif age < 20:
    print("You are a teenager.")
elif age < 60:
    print("You are an adult.")
else:
    print("You are a senior.")
````

**Nested if statements**

````python
age = int(input("Enter your age: "))
has_license = input("Do you have a driver's license? (yes/no): ").lower()

if age >= 18:
    if has_license == "yes":
        print("You can drive cars")
    else:
        print("You need a license to drive cars")
else:
    print("You are too young too drive!")
````

**Using and operator in if statement**

````python
age = int(input("Enter your age: "))
has_license = input("Do you have a driver's license? (yes/no): ").lower()

if age >= 18 and has_license == "yes":
    print("You can drive cars")
else:
    print("You are not allow to drive a car")
````

**Using and operator in if statement**

````python
temperature = int(input("Enter the temperature: "))
is_sunny = input("Is it sunny outside? (yes/no): ").lower()

if temperature > 15 or is_sunny == "yes":
    print("Let's go out.")
else:
    print("It's risky to go outside in this weather condition")
````

## **While Loops**

A loop is used when we want to repeat the same code multiple times without writing it again and again. This makes our code shorter, cleaner, and easier to manage.

A `while` loop repeats a block of code as long as a condition is `True`.

>Think of it as: “Keep doing this `while` something is `True`”

`Basic Structure:`
```python
while condition:# is called loop head
    # code to repeat # is called loop body
```

````python
count = 0 # loop starting

while count < 3: # loop head
    print("Hello World")# loop body
    count = count + 1 # loop update variable
````

## **Infinite `while` loop and loop `break`**

#### **Infinite Loop**
```python
# condition always True
condtion = True
while condition:
    codeblock  # this code block will never stop executing since the condition is always True

````python
num = 12

while True:  # infinite loop
    print(num)
    num = num + 2

    if num > 20:
      break       # stop the loop when condition is met
````


✅ Key Points
- `while True` creates an infinite loop
- `break` is used to exit the loop early
- Always make sure a loop has a stopping condition (or a `break`), otherwise it may run forever
